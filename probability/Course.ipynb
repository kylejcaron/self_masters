## Probability Axioms
---
*Which outcomes are more likely to occur, and which are less likely to occur?*

 * In a continuous model, any individual point should have a probability of zero. Consider trying to hit an exact center of a point with a dart - not very likely, expecially if the point is the size of the darts tip
 * __Instead__ of assigning probabilities to individual points, we assign probabilities to whole sets (a subset of the sample space)
 
__Event__: a subset of the sample space
 * Probability is assigned to events
 * __Example__: A dart falls in the upper half of the dartboard
 
### The Axioms:
 * __Nonnegativity: P(event) >= 0__
     * The probability of the event occurring can't be negative
 * __Normalization: P($\Omega$) = 1__
     * The probability of the entire sample space should always 
     * The outcome is always going to be an element of a sample space, so we have absolute certainty that the outcome is going to be in the sample space
     * We have absolute certainty that event $\Omega$ will occur
 * __(Finite) additivity:__ (to be strengthened later)
     * If $A \cap B = \varnothing$, then $P(A \cup B) = P(A) + P(B) $
     * If the intersection of two sets is empty, then the probability that the outcome of the experiment falls in the union of A and B is simply P(A) + P(B)

## Countable Additivity
If $A_1, A_2, A_3,\dots$ is an infinite sequence of disjoint events, then: 
$$
P(A_1 \cup A_2 \cup A_3 \cup \dots) = P(A_1) + P(A_2) + P(A_3) + \dots 
$$

 * Additivity only holds for "countable" sequences of events
 * The unit square (similarly, the real line, etc.) is not countable (its elements cannot be arranged in a sequence)

__Solution__: Suppose that  𝑐=0 . Then, by countable additivity

$$
1 = P(\Omega) = P(\{1\} \cup \{2\} \cup \{3\} \dots ) = P(\{1\}) + P(\{2\}) + P(\{3\}) + \dots = 0 + 0 + 0 + \dots = 0
$$

which is a contradiction.
